```python
import pandas as pd
import numpy as np

cars = pd.read_csv("cars_dailyRoutes.csv", low_memory=False)
ai4i = pd.read_csv("ai4i2020.csv")
opel = pd.read_csv("opel_corsa_01.csv", sep=';')
```

```python
opel.isnull().sum()
```

```python
opel.fillna(opel.median(numeric_only=True), inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

opel['roadSurface'] = le.fit_transform(opel['roadSurface'])
opel['traffic'] = le.fit_transform(opel['traffic'])
opel['drivingStyle'] = le.fit_transform(opel['drivingStyle'])
```

```python
cols = [
    'roadSurface',
    'traffic',
    'VerticalAcceleration',
    'VehicleSpeedInstantaneous'
]

for c in cols:
    opel[c] = (
        opel[c]
        .astype(str)
        .str.replace(',', '.', regex=False)
    )

for c in cols:
    opel[c] = pd.to_numeric(
        opel[c],
        errors='coerce'
    )

opel[cols] = opel[cols].fillna(
    opel[cols].median()
)

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

opel[cols] = scaler.fit_transform(
    opel[cols]
)
```

```python
print(opel['roadSurface'].unique())
print(opel['traffic'].unique())
```

```python
opel['IRSI'] = (
      0.35 * opel['roadSurface']
    + 0.30 * opel['traffic']
    + 0.20 * abs(opel['VerticalAcceleration'])
    + 0.15 * opel['VehicleSpeedInstantaneous']
) * 100
```

```python
opel['LongitudinalAcceleration'] = (
    opel['LongitudinalAcceleration']
    .astype(str)
    .str.replace(',', '.', regex=False)
)

opel['LongitudinalAcceleration'] = pd.to_numeric(
    opel['LongitudinalAcceleration'],
    errors='coerce'
)
```

```python
opel['LongitudinalAcceleration'].fillna(
    opel['LongitudinalAcceleration'].median(),
    inplace=True
)
```

```python
opel['drivingStyle'].fillna(
    opel['drivingStyle'].mode()[0],
    inplace=True
)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

opel[['LongitudinalAcceleration','drivingStyle']] = (
    scaler.fit_transform(
        opel[['LongitudinalAcceleration','drivingStyle']]
    )
)
```

```python
opel['DrivingAggressiveness'] = (
      0.4 * opel['VehicleSpeedInstantaneous']
    + 0.4 * abs(opel['LongitudinalAcceleration'])
    + 0.2 * opel['drivingStyle']
) * 100
```

```python
opel[['DrivingAggressiveness']].head()
```

```python
opel['DrivingAggressiveness'].describe()
```

```python
opel['FuelConsumptionAverage'].dtype
opel['VehicleSpeedAverage'].dtype
```

```python
for c in ['FuelConsumptionAverage', 'VehicleSpeedAverage']:
    opel[c] = (
        opel[c]
        .astype(str)
        .str.replace(',', '.', regex=False)
    )

    opel[c] = pd.to_numeric(
        opel[c],
        errors='coerce'
    )
```

```python
opel['FuelConsumptionAverage'].fillna(
    opel['FuelConsumptionAverage'].median(),
    inplace=True
)

opel['VehicleSpeedAverage'].fillna(
    opel['VehicleSpeedAverage'].median(),
    inplace=True
)
```

```python
opel['FuelEfficiencyScore'] = (
    opel['VehicleSpeedAverage']
    /
    (opel['FuelConsumptionAverage'] + 1)
)
```

```python
opel[['FuelEfficiencyScore']].head()
```

```python
def road_category(x):
    if x < 30:
        return 'Good'
    elif x < 60:
        return 'Moderate'
    else:
        return 'Bad'

opel['RoadCategory'] = opel['IRSI'].apply(road_category)
```
